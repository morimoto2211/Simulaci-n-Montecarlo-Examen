```python
import pandas as pd
import random
import numpy as np
import math
```

```python
def simulacion_montecarlo(n_iteraciones, n_satelites, num_satelites_necesarios, df=None):
    outputs = []
    if df is None:
        for i in range(n_iteraciones):
            tiempo_falla = np.random.uniform(1000,5000, n_satelites)
            tiempo_falla = np.sort(tiempo_falla)[::-1]
            tiempo_expected_simulacion = tiempo_falla[num_satelites_necesarios-1]
            outputs.append(tiempo_expected_simulacion)
    
    if df is not None:
        for i in range(len(df.columns)):
            tiempos_fallas = df.iloc[:,i].values
            tiempos_fallas = np.sort(tiempos_fallas)[::-1]
            tiempo_expected_simulacion = tiempos_fallas[num_satelites_necesarios-1]
            outputs.append(tiempo_expected_simulacion)
    
    promedio_tiempofalla = round(sum(outputs)/len(outputs),2)
    dsv_est = round(np.std(outputs),2)
    
    return promedio_tiempofalla, dsv_est
```

```python
promedio_simulacion, desv_est = simulacion_montecarlo(1000, 5, 2)
print(promedio_simulacion, desv_est)
```
